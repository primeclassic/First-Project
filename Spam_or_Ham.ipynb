```python
# Deep Learning Frameworks
import tensorflow as tf
import torch
# Data Handling
import pandas as pd
import numpy as np
# Evaluation Metrics
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("mail_data.csv")
df.head(10)
#the dataset consists of 2 columns
```

```python
print(df.isnull().sum())  # Check for missing values
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Message'], df['Category'], test_size= .2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Transform text data into TF-IDF features
# Will use a range of max_features becaues this dataset is quite small range from 2000-5000
max_features_range = [100,1000,2000,2500,3000,3500,4000,4500,5000]
results = []
# Iterate through the range of max features
for max_features in max_features_range:
  # Create TfidVectorizer with our range of max features
  vectorizer = TfidfVectorizer(max_features=max_features)

  # Vectorizing training and testing data
  X_train_tfidf = vectorizer.fit_transform(X_train)
  X_test_tfidf = vectorizer.transform(X_test)

  # Train a model (Naive Bayes)
  model = MultinomialNB()
  model.fit(X_train_tfidf,y_train)
  # Predict on test set
  y_pred = model.predict(X_test_tfidf)

  # Accuracy
  accuracy = accuracy_score(y_test,y_pred)

  # Store accuracy
  results.append((max_features,accuracy))

# Print results
for max_features, accuracy in results:
  print(f"max_features: {max_features}, Accuracy: {accuracy}")
# Find the optimal max_features value
optimal_max_features = max(results, key=lambda item: item[1])[0]
print(f"Optimal max_features: {optimal_max_features}")
```

```python
# I will use max_features of 1000 because it has the highest accuracy and compare it to some real emails
# Create test DataFrame
new_emails = [
    {"Message": "We're excited to invite you to be part of something extraordinary...", "Category": "ham"},
    {"Message": "You've won a free prize! Click here to claim...", "Category": "spam"},
    {"Message": "The Leanpub Weekly Sale for March 7 2025 This is a weekly sale email. You can subscribe or unsubscribe here", "Category": "ham"},
    {"Message": "The Office of Academic Affairs is delighted to announce exciting remote internship opportunities...", "Category": "spam"},
    {"Message": "Check Now W.F BANK. ACC-LOCK: YOU ARE MANDATED TO VER1FY YOUR ACCOUNT BELOW:", "Category": "spam"},
    {"Message": "Hello, I apologize for bothering you via text message. Are you looking for a job?", "Category": "spam"}
]
test_df = pd.DataFrame(new_emails)
# Preprocess and predict
```

```python
# Create TfidVectorizer with our range of max features
vectorizer = TfidfVectorizer(max_features=200)

# Vectorizing training and testing data
X_train_tfidf = vectorizer.fit_transform(X_train)

# Train a model (Naive Bayes)
model = MultinomialNB()
model.fit(X_train_tfidf,y_train)
# Preprocess and predict using the trained model with max_features = 1000
new_emails_tfidf = vectorizer.transform(test_df['Message'])
predictions = model.predict(new_emails_tfidf)

# Accuracy
accuracy = accuracy_score(test_df['Category'],predictions)

print(f"Accuracy on new emails: {accuracy}")
```

```python
# I will try a larger amount of emails instead
import pandas as pd
import csv
import random
# Sample spam and ham email messages
spam_messages = [
    "Congratulations! You've won a free prize. Click here to claim it now!",
    "Your account has been compromised. Please verify your details immediately.",
    "Get a free loan with low interest rates. Apply now!",
    "Limited-time offer: Huge discounts on all products. Shop now!",
    "You've been selected for a free trial. Claim your offer today!",
    "Weight Loss! Limited Time!",
    "Get Rich Quick!",
    "Work From Home",
    "Meet Singles",
    "Debt Relief",
    "You have Won!",
    "Free Gift",
    # All caps messages most likely a spam if you see one
    random.choice([
        "URGENT! LIMITED TIME OFFER! CLAIM YOUR FREE GIFT NOW!",
        "YOU HAVE BEEN SELECTED FOR A FREE TRIP TO THE BAHAMAS!",
        "CONGRATULATIONS! YOU'RE A WINNER!",
    ]),
    # Additional scenarios
    "Your package is delayed due to unforeseen circumstances. Please contact us.",
    "Important security alert for your online banking account.",
    "Your subscription is about to expire. Renew now to avoid disruption."
]

ham_messages = [
    "Hi John, I hope this email finds you well. Let's catch up soon.",
    "Regarding our meeting next week, please find the agenda attached.",
    "Thank you for your order. Your package will be shipped soon.",
    "Reminder: Project deadline is approaching. Please submit your work on time.",
    "Just a friendly reminder about the upcoming event.",
    "I hope you had a wonderful weekend and are ready to get back to work!",
    "Can you please send me the reports by Monday",
    "Thank you for reaching out, please find information at this link",
    "Hello, please reply to schedule a meeting at your earliest convenience",
    "I look forward to our upcoming meeting on",
    # Additional scenarios
    "Quick question about the project.",
    "Let me know if you're available for a call later today.",
    "Just checking in to see how things are going.",
    "Please see the attached document for your review."
]


# Generate data
data = []
for i in range(50):  # 50 spam emails
    data.append([spam_messages[i % len(spam_messages)], "spam"])
for i in range(50):  # 50 ham emails
    data.append([ham_messages[i % len(ham_messages)], "ham"])

# Create DataFrame
df = pd.DataFrame(data, columns=["Message", "Category"])

# Shuffle the rows
shuffled_df = df.sample(frac=1, random_state=42)

# Reset the index
shuffled_df = shuffled_df.reset_index(drop=True)
```

```python
# Train model using new data frame
# Create TfidVectorizer with our range of max features
vectorizer = TfidfVectorizer(max_features=1000)

# Vectorizing training and testing data
X_train_tfidf = vectorizer.fit_transform(X_train)

# Train a model (Naive Bayes)
model = MultinomialNB()
model.fit(X_train_tfidf,y_train)

# Vectorize the smaller dataset
X_shuffled_tfidf = vectorizer.transform(shuffled_df['Message'])

# Make predictions on the smaller dataset
predictions_shuffled = model.predict(X_shuffled_tfidf)

# Evaluate accuracy on the smaller dataset
accuracy_shuffled = accuracy_score(shuffled_df['Category'], predictions_shuffled)

print(f"Accuracy on the smaller dataset (shuffled_df): {accuracy_shuffled}")
```

```python
# Implementing hyperparams
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
# Train model using new data frame
# Create TfidVectorizer with our range of max features
vectorizer = TfidfVectorizer(max_features=1000)

# Vectorizing training and testing data
X_train_tfidf = vectorizer.fit_transform(X_train)

# Train a model (Naive Bayes)
model = MultinomialNB()

# Hyperparameter
param_grid = {
    'alpha': [0.01, 0.1, 1],
    'fit_prior': [True, False]
}

# Grid search
grid_search = GridSearchCV(estimator=model, param_grid = param_grid, cv = 5, scoring = 'accuracy')

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train_tfidf, y_train)

# Get best
best_model = grid_search.best_estimator_   # Use this instead of fitting model

# Vectorize the smaller dataset
X_shuffled_tfidf = vectorizer.transform(shuffled_df['Message'])

# Make predictions on the smaller dataset
predictions_shuffled = best_model.predict(X_shuffled_tfidf)

# Evaluate accuracy on the smaller dataset
accuracy_shuffled = accuracy_score(shuffled_df['Category'], predictions_shuffled)

print(f"Accuracy on the smaller dataset (shuffled_df): {accuracy_shuffled}")
```
